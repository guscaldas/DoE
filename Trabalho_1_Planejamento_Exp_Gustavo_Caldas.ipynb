# 2. Introdução

O uso de adesivos biodegradáveis à base de amido encontram aplicações industriais, sendo uma fonte de matéria-prima considerada renovável. Uma dessas fontes de adesivos industriais é o milho (Zea mays L.), que é comumente processados por meio de tratamentos químicos e térmicos, como a adição de um agente hidrolisante.  Outras fontes comuns de amido incluem tapioca e batata, entre outros. Por isso, há o interesse de obtenção de adesivos à base de produtos que não compitam com fontes de alimentação humana, sendo proposto o adesivo à base de Ramon (Brosimum alicastrum Sw.).

Uma forma de produzir benchmarking na indústria de adesivos é por meio das suas propriedades mecânicas. Testes de resistência ao rasgamento e ao cisalhamento são formas comuns para tal. Esses testes são resultados de uma série de fatores que afetam a físico-química entre o adesivo e o substrato, em que forças coesivas e adesivas são formadas. 

A  qualidade do adesivo, portanto, depende das forças do ligante do polímero usado, que por sua vez é produto dos diversos fatores em que se faz a sua preparação. De forma a obter a melhor combinação possível de um tratamento,o planejamento composto central (CCD) é uma metodologia de planejamento de experimentos usada em conjunto com os métodos de superfície  resposta de resposta que tem como objetivo otimizar um variável de interesse cuja influência é dada a partir de outras variáveis independentes (fatores). Logo, é possível que pontos vulneráveis sejam encontrados na produção e os produtos ou processos se tornem mais robustos. 

A técnica consiste em estender as corridas fatoriais para fora do (hiper)cubo a partir das corridas axiais, por isso é comumente chamado de planejamento em estrela. Dessa forma, é possível obter uma curva de resposta adequada. Comumente, o CCD é uma ampliação de um planejamento inicial. Por exemplo, um planejamento fatorial com pontos centrais poderia ser feito para indicar se a curvatura é significativa. O método do gradiente descendente (ascendente) irá indicar a direção que o planejamento deve ser estendido, onde a partir desse ponto passa-se a um CCD. O tipo de planejamento composto central depende de fatores como a distância do centro até os pontos axiais (alfa). Se alfa é igual a um, é um planejamento cúbico de face centrada; se alfa é igual a raiz quadrada dos fatores, tem-se um (hiper)esfera circunscrita ao (hiper)cubo. Nesses casos, o planejamento é chamado de rotacionável, pois a variância dentro da esfera é a mesma para seus pontos equidistantes.

## 4.1 Análise Exploratória dos Dados - Por que 12 pontos centrais?

Ao analisar o planejamento feito pelos autores, as primeiras perguntas que surgem são: são realmente necessários usar 12 pontos centrais? O que aconteceria se usássemos menos? O modelo de regressão seria muito afetado? Montgomery (2013), por exemplo, sugere 3 a 5 corridas centrais. A justificativa é que no planejamento de experimentos há a preocupação de haver o menor número de experimentos possíveis ao passo que as inferências sejam  com um modelo reduzido. Especialmente nesse caso, estamos tratanto de quatro variáveis (são dois adesivos diferentes e os ensaios são destrutivos), portanto temos na verdade 12 pontos centrais para cada variável, perfazendo um total de 48 experimentos. Qualquer redução que possa ser feita sem comprometer os resultados pode trazer benefícios em termos de tempo para o analista.

Dessa forma, parte-se primeiro para uma análise exploratória dos dados nesses pontos centrais (figuras 3 e 4). Nota-se uma certa variabilidade nos pontos centrais desse experimento, principalmente em termos do adesivo à base de ramon, em que, por exemplo, no caso do  
resistência ao cisalhamento, o intervalo entre o máximo e mínimo estende desde 0.294 a 0.210 MPa e a média está deslocada da mediana. Ainda para o adesivo ramon, a resistência ao rasgamento tem a média levemente deslocada em relação a mediana devido à presença de um  valor máximo que é cuja magnitude da diferença com os limites quartis é maior. 

Os diagramas de caixa, contudo, não indicam nenhum "outlier". Pelo gráfico de quantis normais (à exceção de res. ao rasgamento do milho), todos seguem a mesma tendência. Dessa forma, optou-se por eliminar alguns pontos de forma à investigar se a análise será afetada (pontos 25,32,35,36 na ordem normal).

### 4.2.1 Escolhendo um modelo de curva de resposta

A primeira análise será feito usando todos os termos de segunda ordem para as quatro variáveis tanto para o caso quando se usa todos os pontos centrais quando se exclui. É possível verificar que os modelos de primeira ordem (FO), de interação (TWI) e quadrático (PQ) (tabela 3) são significativos para ambos os casos através da tabela de análise de variância.  

Com 36 experimentos no total (tabela 3) o modelo possui graus de liberdade (GL) de 14, sobrando 21 para os resíduos. Ao retirar 4 pontos centrais, continua-se a ter 14 para os modelos, mas dessa vez, os GLs para os resíduos é de 17. Como o número de graus de liberdade é fixo em 10 para os erros do modelo (numerador), sobram 7 para o erro puro em comparação a 11 com o caso anterior (tabela 4). 

Logo, da tabela de valores críticos de F (não mostrada aqui), o $F_{critic}$ passa de 2.85 para 3.64 e naturalmente o "lack-of-fit" é afetado. Conforme a tabela 4 mostra, ele é significativo (p<0.05). Um outro argumento é o aumento dos erros-padrão em alguns coeficientes (tabelas 3 e 5). Esse modelo portanto é rejeitado e o modelo original é mantido. Portanto, optou-se por não excluir nenhum ponto central. 

Se diminuíssimos ainda mais o número de corridas axiais, verificar-se-ia que $F_{critic}$ passaria a aumentar ainda mais e, portanto, o teste de "lack-of-fit" teria pouca capacidade em baixo número de corridas centrais, mesmo quando há pouca adequação do modelo. Como o ponto central é o que queremos que tenha a menor variância possível, um número pequeno de corridas centrais frente às corridas fatoriais pode incorrer em uma variância pouco estável dentro da região da esfera.

O uso de metodologias de superfícies de resposta mostra-se como uma ferramenta no melhoramento da qualidade de produtos e processos, através da fácil visualização de pontos ótimos.
```r
#Função para ler um modelo e visualizar os p-values em vermelho
anova.table  <-     function(lm.model){
                    anova_table <- tidy(anova(lm.model))
                    anova_table <- anova_table %>% kable("html") %>%
                    row_spec(which(anova_table$p.value < 0.05), bold = T, color = "red") %>%
                    as.character() #%>%
                    #display_markdown("**Análise da Variância - Médias**")
                    display_html(anova_table)
                                 
                    }
lof.table  <-     function(rsm.model){
                    lof <- tidy(summary(rsm.model)$lof)
                    lof <- lof %>% kable("html") %>% row_spec(which(lof$p.value < 0.05), bold = T, color = "red") %>% as.character() #%>%
                    #display_markdown("**Análise da Variância - Lack-of-fit**")
                    display_html(lof)  
                    }
coef.table  <-     function(lm.model){
                    reg <- summary(lm.model)$coefficients
                    reg <- reg %>% kable("html") %>% row_spec(which(reg[,"Pr(>|t|)"] < 0.05), bold = T, color = "red") %>% as.character() #%>%
                    #display_markdown("**Coeficientes de regressão**")
                    display_html(reg)  
                    }
#Função R^2
r.quad <-   function(rsm.model){
                cat('R^2:',summary(rsm.model)$r.squared,"\n")
                cat('R^2 ajustado:',summary(rsm.model)$adj.r.squared)
}
```
